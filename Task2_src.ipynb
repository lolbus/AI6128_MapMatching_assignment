```python
import matplotlib.pyplot as plt
import osmnx as ox
from descartes import PolygonPatch
from shapely.geometry import MultiPolygon
from shapely.geometry import Polygon
import pandas
import ast
import numpy as np

data = pandas.read_csv('./data/train_1000.csv')
```

```python
#list(data.iloc[0][-1])
#list_of_points = ast.literal_eval(data.iloc[0][-1])
def correct_annotate_points_valid(point, text, maxX):
    x = point[0]
    y = point[1]
    textlen = len(text)
    minX_req = textlen * 0.00030 - 0.0012
    if (maxX - x < minX_req): #shift to left if too right
        correction_factor = textlen * 0.00030 - (maxX - x) - 0.0012
        x = x - correction_factor
    return (x, y)

for trip in range(0,10):
    coords = np.array(ast.literal_eval(data.iloc[trip]['POLYLINE']))
    list_of_x = coords[:, 0]
    list_of_y = coords[:, 1]
    min_x = min(list_of_x - 0.005)
    max_x = max(list_of_x + 0.005)
    min_y = min(list_of_y - 0.005)
    max_y = max(list_of_y + 0.005)
    boundary = [min_x, max_x, min_y, max_y]
    boundary_polygon = Polygon([(min_x,min_y),(max_x,min_y),(max_x,max_y),(min_x,max_y)])
    G = ox.graph_from_polygon(boundary_polygon, network_type='drive')
    #nc = ox.plot.get_node_colors_by_attr(G, attr="y", cmap="viridis")
    fig, ax = ox.plot_graph(
        G,
        show=False,
        close=False,
        bgcolor="#333333",
        edge_color="w",
        edge_linewidth=0.3,
        node_size=0,
    )
    #plot non labeled
    plt.plot(list_of_x,list_of_y,c="r", linewidth=5, alpha = 0.5)
    plt.plot(list_of_x[0],list_of_y[0],c="r", marker ="o")
    
    plt.annotate("TrajectoryID:"+str(trip), (min_x + 0.005, max_y - 0.0015), c='w', fontsize=18)
    plt.show()
    
    
    
    #Plot labeled graph
    G = ox.graph_from_polygon(boundary_polygon, network_type='drive')
    #nc = ox.plot.get_node_colors_by_attr(G, attr="y", cmap="viridis")
    fig, ax = ox.plot_graph(
        G,
        show=False,
        close=False,
        bgcolor="#333333",
        edge_color="w",
        edge_linewidth=0.3,
        node_size=0,
    )
    plt.plot(list_of_x,list_of_y,c="r", linewidth=5, alpha = 0.5)
    plt.plot(list_of_x[0],list_of_y[0],c="r", marker ="o")
    
    plt.annotate("TrajectoryID:"+str(trip), (min_x + 0.005, max_y - 0.0015), c='w', fontsize=18)
    text_start = 'Start' + str((list_of_x[0], list_of_y[0]))
    text_end = 'End' + str((list_of_x[len(list_of_x)-1], list_of_y[len(list_of_x)-1]))
    plt.plot(list_of_x[len(list_of_x)-1],list_of_y[len(list_of_y)-1],c="r", marker ="o")
    annotate_point_start = correct_annotate_points_valid((list_of_x[0], list_of_y[0]), text_start, max_x)
    
    plt.annotate(text_start, (annotate_point_start[0],annotate_point_start[1]), c='r', fontsize=12)
    plt.annotate(text_end, (list_of_x[len(list_of_x)-1]+0.0005,list_of_y[len(list_of_y)-1]+0.0005), c='r', fontsize=12)
    #plt.savefig('unmatched_map_'+str(trip)+'.png')
    plt.show()
    print()
    
    #5, 8, 6, 2, 3, 10, 9, 1, 4, 7 
```
