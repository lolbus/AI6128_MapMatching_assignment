### Load essential libraries and post-proceesed data of FMM

```python
import pandas
import numpy as np
import osmnx as ox
import time
from shapely.geometry import Polygon
import os
import ast
import matplotlib.pyplot as plt

fmm_output = pandas.read_csv('postprocessed_output.csv')
```

```python
fmm_output = fmm_output.sort_values(by='trajectory_id', ascending=True)
fmm_output[0:10]
```

```python
len(fmm_output)
```

### Generate 10 Figures from the first 10 mapped routes

```python

image_to_visualize = 10
image_visualized = 0
for i, row in enumerate(range(len(fmm_output))):  
    print("Processing.. ", i)
    trajectoryID = i
    idx = fmm_output.iloc[i:i+1]['fmm-mapped WSG84 points sequence'].index[0]
    ms_stat = fmm_output.iloc[i:i+1]['match_success'].bool()
    if ms_stat == False:
        print("dropping trajectory idx ", trajectoryID)
        continue
    if image_visualized == 10:
        print("END")
        break  
    image_visualized += 1
    #print("plotting image: ", image_visualized)
    print("row index", idx)
    print("trajectory index", trajectoryID)
    #fmm
    fmm = np.array(ast.literal_eval(fmm_output.iloc[i:i+1]['fmm-mapped WSG84 points sequence'][idx]))

    # raw
    raw_coords = np.array(ast.literal_eval(fmm_output.iloc[i:i+1]['raw_gps_trajectories'][idx]))
    raw_list_of_x = raw_coords[:, 0]
    raw_list_of_y = raw_coords[:, 1]

    # FMM-ed
    coords = fmm
    list_of_x = coords[:, 0]
    list_of_y = coords[:, 1]
    
    min_x = min(list_of_x - 0.005)
    max_x = max(list_of_x + 0.005)
    min_y = min(list_of_y - 0.005)
    max_y = max(list_of_y + 0.005)
    boundary = [min_x, max_x, min_y, max_y]
    boundary_polygon = Polygon([(min_x,min_y),(max_x,min_y),(max_x,max_y),(min_x,max_y)])
    G = ox.graph_from_polygon(boundary_polygon, network_type='drive')
    fig, ax = ox.plot_graph(
        G,
        show=False,
        close=False,
        bgcolor="#333333",
        edge_color="w",
        edge_linewidth=0.3,
        node_size=0,
    )
    

    print('raw start')
    print(raw_list_of_x[0] , ',' , raw_list_of_y[0])
    print('raw end')
    print(raw_list_of_x[-1] , ',' , raw_list_of_y[-1])
    
    print('mapped start')
    print(list_of_x[0] , ',' , list_of_y[0])
    print('mapped end')
    print(list_of_x[-1] , ',' , list_of_y[-1])
    
    
    #plot route
    plt.plot(raw_list_of_x,raw_list_of_y,c="r", linewidth=5, alpha = 0.7)
    plt.plot(list_of_x,list_of_y,c="b", linewidth=5, alpha = 0.7)
    plt.plot(list_of_x[0],list_of_y[0],c="green", marker ="o", markersize = 12)
    plt.plot(list_of_x[len(list_of_x)-1],list_of_y[len(list_of_y)-1],c="b", marker ="o")
    #plot raw route
    plt.plot(raw_list_of_x[0],raw_list_of_y[0],c="hotpink", marker ="o")
    plt.plot(raw_list_of_x[-1],raw_list_of_y[-1],c="r", marker ="o")


    import matplotlib.patches as mpatches
    red_patch = mpatches.Patch(color='red', label='Raw data')
    blue_patch = mpatches.Patch(color='blue', label='Mapped data')
    #ax.legend(handles=[red_patch])
    plt.legend(handles=[red_patch, blue_patch], loc ="lower right")
    #plt.annotate('Start', (list_of_x[0]+0.0005,list_of_y[0]+0.0005), c='r')
    #plt.annotate('End', (list_of_x[len(list_of_x)-1]+0.0005,list_of_y[len(list_of_y)-1]+0.0005), c='r')
    plt.show()

```
